```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/pookay/Desktop/German_Credit_risk_analysis/dataset/german_credit_data.csv')
```

```python
df.info()
```

```python
print("First 5 rows of the dataset")
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['Saving accounts'].fillna('none', inplace=True)
df['Checking account'].fillna('none', inplace=True)
```

```python
df.info()
```

```python
purpose_risk = df.groupby('Purpose')['Risk'].value_counts(normalize=True).unstack().fillna(0)
purpose_risk['default_rate'] = purpose_risk['bad'] * 100

purpose_risk = purpose_risk.sort_values(by='default_rate', ascending=False)

print("Default Rate by Loan Purpose ")
purpose_risk
```

```python
# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=purpose_risk['default_rate'], y=purpose_risk.index, palette='viridis')
plt.title('Default Rate by Loan Purpose', fontsize=16)
plt.xlabel('Default Rate (%)', fontsize=12)
plt.ylabel('Purpose of Loan', fontsize=12)
plt.show()
```

```python
housing_risk = df.groupby('Housing') ['Risk'].value_counts(normalize=True).unstack().fillna(0)
housing_risk['default_risk']=housing_risk['bad']*100

housing_risk = housing_risk.sort_values(by='default_risk', ascending=False)

print("Default rate by Housing Type")
housing_risk
```

```python
# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=housing_risk['default_risk'], y=housing_risk.index, palette='magma')
plt.title('Default Rate by Housing Type', fontsize=16)
plt.xlabel('Default Rate (%)', fontsize=12)
plt.ylabel('Housing Type', fontsize=12)
plt.show()
```

```python

categorical_cols = ['Sex', 'Housing', 'Saving accounts', 'Checking account', 'Purpose']

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

for col, ax in zip(categorical_cols, axes.flatten()):
    risk_analysis = df.groupby(col)['Risk'].value_counts(normalize=True).unstack().fillna(0)
    risk_analysis['default_rate'] = risk_analysis['bad'] * 100

    sns.barplot(x=risk_analysis['default_rate'], y=risk_analysis.index, ax=ax, palette='viridis')
    ax.set_title(f'Default Rate by {col}', fontsize=14)
    ax.set_xlabel('Default Rate (%)')
    ax.set_ylabel('')
fig.delaxes(axes[2, 1])    
plt.tight_layout()
plt.show()
```
